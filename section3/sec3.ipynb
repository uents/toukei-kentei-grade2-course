# §3. 確率統計

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from scipy import stats
import plotly.io as pio

# VS Code Notebook用のレンダラーを設定
pio.renderers.default = 'plotly_mimetype+notebook'
```

## 歪度・尖度の実装例

```python
# データの生成
np.random.seed(42)

# 異なる分布のデータを生成
normal_data = np.random.normal(0, 1, 10000)  # 正規分布
right_skewed = np.random.exponential(2, 10000)  # 右に歪んだ分布
left_skewed = -np.random.exponential(2, 10000)  # 左に歪んだ分布
high_kurtosis = np.random.laplace(0, 1/np.sqrt(2), 10000)  # 高い尖度（ラプラス分布）

# サブプロットを作成（1行3列）
fig = make_subplots(
    rows=3, cols=1,
    subplot_titles=[
        '正の歪度（右歪み）',
        '負の歪度（左歪み）',
        '高い尖度'
    ]
)

# 比較用のデータセット
comparison_datasets = [
    (right_skewed, '正の歪度（右歪み）', 1, 1, 100),
    (left_skewed, '負の歪度（左歪み）', 2, 1, 100),
    (high_kurtosis, '高い尖度', 3, 1, 100)
]

# 各データセットのヒストグラムを追加
for data, title, row, col, nbins in comparison_datasets:
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    
    # 対象データのヒストグラム
    fig.add_trace(
        go.Histogram(
            x=data,
            nbinsx=nbins,
            name=title,
            histnorm='probability density',
            marker=dict(
                color='rgba(0, 100, 200, 0.7)',
                line=dict(color='black', width=1)
            ),
            showlegend=True,
            legendgroup=f'group{col}'
        ),
        row=row, col=col
    )
    
    # 正規分布を重畳
    fig.add_trace(
        go.Histogram(
            x=normal_data,
            nbinsx=nbins,
            name='正規分布',
            histnorm='probability density',
            marker=dict(
                color='rgba(255, 100, 100, 0.3)',
                line=dict(color='red', width=1)
            ),
            showlegend=True if row == 1 else False,
            legendgroup='normal'
        ),
        row=row, col=col
    )
    
    # 歪度と尖度の情報を注釈として追加
    fig.add_annotation(
        text=f'{title}<br>歪度: {skewness:.3f}<br>尖度: {kurtosis:.3f}',
        xref=f'x{row}', yref=f'y{row}',
        x=0.95, y=0.95,
        xanchor='right', yanchor='top',
        showarrow=False,
        bgcolor='rgba(255, 255, 255, 0.9)',
        bordercolor='black',
        borderwidth=1,
        font=dict(size=10)
    )

# レイアウトを設定
fig.update_xaxes(title_text='値', showgrid=True, gridcolor='rgba(128, 128, 128, 0.3)')
fig.update_yaxes(title_text='密度', showgrid=True, gridcolor='rgba(128, 128, 128, 0.3)')

fig.update_layout(
    height=1400,
    width=800,
    title_text='正規分布との比較による歪度・尖度の可視化',
    showlegend=True,
    barmode='overlay'
)

# グラフを表示
fig.show()

# 統計情報を表示
print("統計情報:")
print("正規分布（参照）:")
print(f"  歪度 (Skewness): {stats.skew(normal_data):.3f}")
print(f"  尖度 (Kurtosis): {stats.kurtosis(normal_data):.3f}")

for data, title, _, _, _ in comparison_datasets:
    print(f"\n{title}:")
    print(f"  歪度 (Skewness): {stats.skew(data):.3f}")
    print(f"  尖度 (Kurtosis): {stats.kurtosis(data):.3f}")
```
